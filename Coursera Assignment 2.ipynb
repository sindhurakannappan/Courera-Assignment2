```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Extracting GameStop stock data using yfinance
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period="max")

# URL of the webpage containing GameStop's revenue data
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Sending a request to fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extracting the revenue data from the table
tables = soup.find_all('table')
revenue_table = None
for table in tables:
    if 'GameStop Quarterly Revenue' in str(table):
        revenue_table = table
        break

# Checking if revenue_table is found
if revenue_table:
    # Converting the HTML table into a pandas DataFrame
    revenue_data = pd.read_html(str(revenue_table))[0]

    # Cleaning up the DataFrame (removing any commas and converting to numeric)
    revenue_data['GameStop Quarterly Revenue(Millions of US $)'] = revenue_data['GameStop Quarterly Revenue(Millions of US $)'].str.replace(',', '').astype(float)

    # Plotting GameStop stock and revenue data
    plt.figure(figsize=(14, 7))

    # Plotting stock price
    plt.subplot(2, 1, 1)
    plt.plot(gamestop_stock_data.index, gamestop_stock_data['Close'], label='Stock Price', color='blue')
    plt.title('GameStop Stock Price and Revenue Over Time')
    plt.ylabel('Stock Price')
    plt.legend()

    # Plotting revenue
    plt.subplot(2, 1, 2)
    plt.plot(revenue_data['Date'], revenue_data['GameStop Quarterly Revenue(Millions of US $)'], label='Revenue', color='orange')
    plt.xlabel('Date')
    plt.ylabel('Revenue (Millions of US $)')
    plt.legend()

    plt.show()
else:
    print("No revenue data found for GameStop.")

```
